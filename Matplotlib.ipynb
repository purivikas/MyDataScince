```python
plt.figure()
# create two panels and set axis
plt.subplot(2,3,1) # (rows,column,panel number)
plt.plot(x,np.sin(x))

# second panel
plt.subplot(2,3,2)
plt.plot(x,np.cos(x))


# second panel
plt.subplot(2,3,3)
plt.plot(x,np.cos(x)+np.sin(x))


plt.subplot(2,3,1) # (rows,column,panel number)
plt.plot(x,np.sin(x)+1)
```

```python
# 0bject Oriented sybtax
# create grid of plot ,ax will be an array of two axes objects

fig, ax = plt.subplots(3)

# Call plot() method on the appropriate object

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x,np.cos(x)+np.sin(x))
ax[1].plot(x,np.cos(x)+1 )
```

```python
# Mt=atlab syntax is more practicticla for quick visualization , but for more complicated visualization
# its stateful interface makes it easy to run into problem , for example one a panel is created , you cant go to 
# the previous one to modify it. therefore the OO interface is considered more standard for complicated visualization.
```

```python
# The Main diffrence between plt.scatter and plt.plot is that it can be used to create scatter plots where the 
# properties of each individual point ( size , face color, edge color, etc) can be individually controlled
# or mapped to data
```

```python
rng = np.random.RandomState(0)
x  = rng.randn(100)
y  = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x,y, c=colors, s=sizes , alpha=0.9, cmap='viridis')
plt.colorbar();
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris  = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.7, s=100*features[3], c=iris.target, cmap='inferno')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
